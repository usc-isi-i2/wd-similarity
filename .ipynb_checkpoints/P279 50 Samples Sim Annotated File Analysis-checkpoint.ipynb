# Language Models Utility Functions

```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
from time import time
import pandas as pd

def getBERTSentEmbeddings(valSeries, modelName):
    model = SentenceTransformer(modelName)
    start = time()
    encodings = model.encode(valSeries.to_list())
    print(time()-start,'s')
    return encodings
```

```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer
model = GPT2LMHeadModel.from_pretrained('gpt2')
tokenizer = GPT2Tokenizer.from_pretrained('gpt2')

def getGPT2Score(sent1):
    # input sentence tensor
    # output score
    tokens_tensor = tokenizer.encode(sent1, add_special_tokens=False, return_tensors="pt")
    loss = model(tokens_tensor, labels=tokens_tensor)[0]
#     print(tokens_tensor.size(1))
    return np.exp(loss.cpu().detach().numpy() / tokens_tensor.size(1))
```

```python
from pytorch_pretrained_bert import BertTokenizer,BertForMaskedLM
import torch
import pandas as pd
import math

bertMaskedLM = BertForMaskedLM.from_pretrained('bert-base-uncased')
berttokenizer = BertTokenizer.from_pretrained('bert-base-uncased')

def getBERTScore(sentence):
    tokenize_input = berttokenizer.tokenize(sentence)
    tensor_input = torch.tensor([berttokenizer.convert_tokens_to_ids(tokenize_input)])
    predictions=bertMaskedLM(tensor_input)
    loss_fct = torch.nn.CrossEntropyLoss()
    loss = loss_fct(predictions.squeeze(),tensor_input.squeeze()).data 
    return math.exp(loss / tensor_input.size(1))
```

# Analysis

## ChildPar

```python
p279_ChildParDF.head()
```

```python
p279_ChildParDF['sent'] = p279_ChildParDF.apply(lambda p: p['node1_label'] + ' is ' + p['node2_label'], axis=1)
```

```python
p279_ChildParDF['node1_emb'] = pd.Series(list(getBERTSentEmbeddings(p279_ChildParDF.node1_label, 'bert-base-nli-mean-tokens')))
p279_ChildParDF['node2_emb'] = pd.Series(list(getBERTSentEmbeddings(p279_ChildParDF.node2_label, 'bert-base-nli-mean-tokens')))
p279_ChildParDF['bert2SentSim'] = p279_ChildParDF.apply(lambda p: cosine_similarity(p.node1_emb.reshape(1,-1), p.node2_emb.reshape(1,-1))[0][0], axis=1)
p279_ChildParDF['bert2SentSim_normed'] = p279_ChildParDF['bert2SentSim'].apply(lambda p: 4 - 3 * p)
```

```python
p279_ChildParDF['gpt2Ppl'] = p279_ChildParDF.sent.apply(getGPT2Score)
p279_ChildParDF['gpt2Ppl_normed'] = p279_ChildParDF['gpt2Ppl'].apply(lambda p: 4 - 3/(1+p))
```

```python
p279_ChildParDF['bert_ppl'] = p279_ChildParDF.sent.apply(getBERTScore)
p279_ChildParDF['bert_ppl_normed'] = p279_ChildParDF['bert_ppl'].apply(lambda p: 4 - 3 / (1+p))
```

```python
p279_ChildParDF.describe()
```
