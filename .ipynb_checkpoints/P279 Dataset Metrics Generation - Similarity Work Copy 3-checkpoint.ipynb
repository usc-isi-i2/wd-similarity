```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
from time import time
import pandas as pd

def getBERTSentEmbeddings(valSeries, modelName):
    model = SentenceTransformer(modelName)
    start = time()
    encodings = model.encode(valSeries.to_list())
    print(time()-start,'s')
    return encodings
```

```python
# abstractsDF_subset = abstractsDF_subset.reset_index()
```

```python
abstractsDF_subset['BERT_abstract_emb'] = pd.Series(list(getBERTSentEmbeddings(abstractsDF_subset.abstract, 'bert-base-nli-mean-tokens')))
```

```python
abstractsDF_subset['BERT_abstract_firstSent_emb'] = pd.Series(list(getBERTSentEmbeddings(abstractsDF_subset.abstract_firstSent, 'bert-base-nli-mean-tokens')))
```

# Analyze 4 Metrics for Cosine Similarity

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
p279WordSimSeededDF_wabs_text = pd.read_csv("../data/P279_dataset/P279_ChildPar_19k_WEmb.csv")
p279Seeded_SiblingsDF3_wabs_text = pd.read_csv("../data/P279_dataset/P279_Siblings_19k_WEmb.csv")
```

```python
p279WordSimSeededDF_wabs_text['textEmb_7props_left'] = p279WordSimSeededDF_wabs_text['textEmb_7props_left'].apply(lambda p: [float(p1) for p1 in p.split(',')])
p279WordSimSeededDF_wabs_text['textEmb_7props_right'] = p279WordSimSeededDF_wabs_text['textEmb_7props_right'].apply(lambda p: [float(p1) for p1 in p.split(',')])
p279WordSimSeededDF_wabs_text['textEmb_2props_left'] = p279WordSimSeededDF_wabs_text['textEmb_2props_left'].apply(lambda p: [float(p1) for p1 in p.split(',')])
p279WordSimSeededDF_wabs_text['textEmb_2props_right'] = p279WordSimSeededDF_wabs_text['textEmb_2props_right'].apply(lambda p: [float(p1) for p1 in p.split(',')])

p279WordSimSeededDF_wabs_text['textEmb_7props_cosSim'] = p279WordSimSeededDF_wabs_text.apply(lambda p: cosine_similarity(np.array(p['textEmb_7props_left']).reshape(1,-1), np.array(p['textEmb_7props_right']).reshape(1,-1)), axis=1)
p279WordSimSeededDF_wabs_text['textEmb_2props_cosSim'] = p279WordSimSeededDF_wabs_text.apply(lambda p: cosine_similarity(np.array(p['textEmb_2props_left']).reshape(1,-1), np.array(p['textEmb_2props_right']).reshape(1,-1)), axis=1)
```

```python
p279WordSimSeededDF_wabs_text['BERT_abstract_emb'] = p279WordSimSeededDF_wabs_text['BERT_abstract_emb'].apply(lambda p: [float(p1) for p1 in p[1:-1].split(',')] if not(pd.isna(p)) else p)
p279WordSimSeededDF_wabs_text['BERT_abstract_emb_right'] = p279WordSimSeededDF_wabs_text['BERT_abstract_emb_right'].apply(lambda p: [float(p1) for p1 in p[1:-1].split(',')] if not(pd.isna(p)) else p)
p279WordSimSeededDF_wabs_text['BERT_abstract_firstSent_emb'] = p279WordSimSeededDF_wabs_text['BERT_abstract_firstSent_emb'].apply(lambda p: [float(p1) for p1 in p[1:-1].split(',')] if not(pd.isna(p)) else p)
p279WordSimSeededDF_wabs_text['BERT_abstract_firstSent_emb_right'] = p279WordSimSeededDF_wabs_text['BERT_abstract_firstSent_emb_right'].apply(lambda p: [float(p1) for p1 in p[1:-1].split(',')] if not(pd.isna(p)) else p)

p279WordSimSeededDF_wabs_text['BERT_abstract_cosSim'] = p279WordSimSeededDF_wabs_text.apply(lambda p: cosine_similarity(np.array(p['BERT_abstract_emb']).reshape(1,-1), np.array(p['BERT_abstract_emb_right']).reshape(1,-1)) if type(p['BERT_abstract_emb']) == list and type(p['BERT_abstract_emb_right']) == list else -1, axis=1)
p279WordSimSeededDF_wabs_text['BERT_abstract_firstSent_cosSim'] = p279WordSimSeededDF_wabs_text.apply(lambda p: cosine_similarity(np.array(p['BERT_abstract_firstSent_emb']).reshape(1,-1), np.array(p['BERT_abstract_firstSent_emb_right']).reshape(1,-1)) if type(p['BERT_abstract_firstSent_emb']) == list and type(p['BERT_abstract_firstSent_emb_right']) == list else -1, axis=1)
```

### Data Split

```python
features = ['textEmb_7props_cosSim', 'textEmb_2props_cosSim', 'BERT_abstract_cosSim', 'BERT_abstract_firstSent_cosSim']
```

```python
from sklearn.model_selection import KFold
X = p279WordSimSeededDF_wabs_text[features]
Y = p279WordSimSeededDF_wabs_text['bert2SentSim']
```

```python
N_SPLITS = 10
```

```python
skf = KFold(n_splits=N_SPLITS, random_state=19, shuffle=True)
X_train_splits, X_test_splits, Y_train_splits, Y_test_splits = [], [], [], []
for train_index, test_index in skf.split(X, Y):
    X_train_splits.append(X.iloc[train_index])
    X_test_splits.append(X.iloc[test_index])
    Y_train_splits.append(Y.iloc[train_index])
    Y_test_splits.append(Y.iloc[test_index])
```

```python
X_train_splits[0].head()
```

### SVM Training and Testing

```python
import numpy as np
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVR

from sklearn.metrics import mean_squared_error, explained_variance_score
import seaborn as sns
from sklearn.metrics import plot_confusion_matrix
import matplotlib.pyplot as plt 
```

```python
preds = []
for X_train1, Y_train1, X_test1, Y_test1 in zip(X_train_splits, Y_train_splits, X_test_splits, Y_test_splits):
    clf = make_pipeline(StandardScaler(), SVR(gamma='auto'))
    clf.fit(X_train1, Y_train1)
    preds.append(clf.predict(X_test1))
```

### Results

```python
acc = 0
for pred, Y_test1 in zip(preds, Y_test_splits):
    acc += mean_squared_error(pred, Y_test1)
acc/N_SPLITS
```

```python
indNames = ['Text Embedding - 7 props', 'Text Embedding - P279 + P31', 'DBPedia short abstracts', 'DBPedia short abstracts first sentence']
```

```python
tempVals = []
for col in features:
    preds = []
    for X_train1, Y_train1, X_test1, Y_test1 in zip(X_train_splits, Y_train_splits, X_test_splits, Y_test_splits):
        clf1 = make_pipeline(StandardScaler(), SVR(gamma='auto'))
        clf1.fit(X_train1[[col]], Y_train1)
        preds.append(clf1.predict(X_test1[[col]]))
    acc = 0
    for pred, Y_test1 in zip(preds, Y_test_splits):
        acc += mean_squared_error(pred, Y_test1)
    tempVals.append(acc/N_SPLITS)
```

```python
(pd.DataFrame(tempVals, index = indNames, columns = ['MSE compared with BERT 2 label emb score']) * 100).sort_values(by=['MSE compared with BERT 2 label emb score'], ascending=False)
```
