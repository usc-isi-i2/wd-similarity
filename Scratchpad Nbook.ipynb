```python
wordSim353AnnotDF_New = pd.read_csv('../data/wordsim353LatestAnnot.csv')
wordSimOG_DF = pd.read_csv('../data/wordsim.csv')
```

```python
wordSim353AnnotDF_New = wordSim353AnnotDF_New.set_index(['Word 1', 'Word 2']).join(wordSimOG_DF.set_index(['Word 1', 'Word 2'])[['word1_kg_id', 'word2_kg_id']]).reset_index()
```

```python
wordSim353AnnotDF_New = wordSim353AnnotDF_New[['word1_kg_id', 'word2_kg_id', 'Avg']]
temp1 = wordSim353AnnotDF_New_Merged_DF.set_index(['word1_kg_id', 'word2_kg_id']).join(wordSim353AnnotDF_New.set_index(['word1_kg_id', 'word2_kg_id']), rsuffix="_old").reset_index()
```

```python
temp2 = temp1[temp1.Avg != temp1.Avg_old][['Avg', 'Avg_old']]
```

```python
def labelSamples(score):
    return 'I' if score <= 1.75 else 'U' if score >= 3.5 else 'M'
```

```python
temp2['Cat'] = temp2.Avg.apply(labelSamples)
temp2['Cat_old'] = temp2.Avg_old.apply(labelSamples)
temp2
```
