```python
import pandas as pd
```

```python
def findMaximals(df):
    embDict = {}
    for _, row in df.iterrows():
        emb = row['Embedding']
        basis = row['Basis']
        newAcc = row['New Acc']
        oldAcc = row['Old Acc']
        iteration = row['Iteration Num']
        if emb not in embDict:
            embDict[emb] = {'maxNewAcc': newAcc, 'maxRow': row, 'oldAcc': -1}
        else:
            if newAcc > embDict[emb]['maxNewAcc'] or (newAcc == embDict[emb]['maxNewAcc'] and 'class' in basis):
                embDict[emb] = {'maxNewAcc': newAcc, 'maxRow': row, 'oldAcc': embDict[emb]['oldAcc']}
        if iteration == 1:
            embDict[emb]['oldAcc'] = oldAcc
    finalRows = []
    for emb in embDict:
        finalRows.append([*embDict[emb]['maxRow'].to_list(), embDict[emb]['oldAcc'], embDict[emb]['maxRow']['New Acc'] - embDict[emb]['oldAcc']])
    return pd.DataFrame(finalRows)
```

```python
unsupSumm = pd.read_csv('../data/retrofitting/retro_test_Sept_21_2021.csv')
findMaximals(unsupSumm)
```

```python
def findSVMMaximals(df):
    embDict = {}
    for _, row in df.iterrows():
        emb = row['emb']
        basis = row['basis']
        acc = row['Acc']
        iteration = row['iteration']
        if iteration == 0:
            if emb not in embDict:
                embDict[emb] = {'maxNewAcc': -1, 'maxRow': row, 'oldAcc': acc}
            else:
                embDict[emb] = {'maxNewAcc': embDict[emb]['maxNewAcc'], 'maxRow': embDict[emb]['maxRow'], 'oldAcc': acc}
        else:
            if emb not in embDict:
                embDict[emb] = {'maxNewAcc': acc, 'maxRow': row, 'oldAcc': -1}
            else:
                if acc > embDict[emb]['maxNewAcc'] or (acc == embDict[emb]['maxNewAcc'] and 'class' in basis):
                    embDict[emb] = {'maxNewAcc': acc, 'maxRow': row, 'oldAcc': embDict[emb]['oldAcc']}
    finalRows = []
    for emb in embDict:
        finalRows.append([*embDict[emb]['maxRow'].to_list(), embDict[emb]['oldAcc'], embDict[emb]['maxRow']['Acc'] - embDict[emb]['oldAcc']])
    return pd.DataFrame(finalRows)
```

```python
svmSumm = pd.read_csv('../data/retrofitting/retro_SVM_Reg_test_Sept_21_2021.csv')
findSVMMaximals(svmSumm)
```
